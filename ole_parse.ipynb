### Compare sizes

```python
def get_sl(file_path):
    with olefile.OleFileIO(file_path) as ole:
        streams = ole.listdir(streams=True, storages=False)    
        lengths = [ole.get_size(s) for s in streams]

        sl = sorted(zip(streams, lengths), key=lambda t: -t[1])
        
    return sl

sl1 = get_sl("data/Project1/2021/09_September/20210922/20210922 - LC_22092021_001.lcd")
sl2 = get_sl("data/Project1/2021/09_September/20210908/20210908_08092021_003.lcd")

both = [t for t in sl1 if not t in sl2]
only_sl1 = [t for t in sl1 if not t in both]
only_sl2 = [t for t in sl2 if not t in both]
```
